```python
from gensim.models import Word2Vec, KeyedVectors
import pandas as pd
import nltk
```

```python
df = pd.read_csv("all_seasons.csv")
```

```python
df.head()
```

```python
newTitles = df['college'].values
```

```python
newTitles
```

```python
#nltk.download('punkt')
```

```python
newsVec = [nltk.word_tokenize(college) for college in newTitles]
newsVec
```

```python
model = Word2Vec(newsVec,min_count = 1,vector_size=32)
```

```python
model.wv.most_similar(positive=['Stanford'])
```

```python
vec = model.wv.get_vector('Lamar') - model.wv.get_vector('Xavier') + model.wv.get_vector('Alabama')

model.wv.most_similar(positive=[vec])
```
